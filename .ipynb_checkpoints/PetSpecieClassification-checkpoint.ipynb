# Importing the dataset
We'll use untar function to find the tar file of the dataset in the provided url and then extract the actual data.

Let us now use the untar function on this url.

```python
path = untar_data(URLs.PETS);
path
```

```python
path.ls()
```

```python
annotationsFolder = path/'annotations'
imagesFolder = path/'images'
```

Let us now see what exactly our dataset looks like.

```python
fileNames = get_image_files(imagesFolder)
```

```python
fileNames[:10]
```

As we can see, the data images have their number and their label in their filenames. We need the labels in order to check the classification of our model. Therefore, we need some way to extract these labels.

We might be able to accomplish this via a regex.


```python
regexPattern = r'/([^/]+)_\d+.jpg$'
```

```python
data = ImageDataBunch.from_name_re(imagesFolder, fileNames, regexPattern, ds_tfms = get_transforms(), size=224)
data.normalize(imagenet_stats)
```

Let's see how our images look.

```python
data.show_batch(rows=2, figsize=(7,6))
```

 The classes available that our model needs to classify against.

```python
data.classes
```

```python
len(data.classes)
```

# Training the dataset
We will be using a convolutional nueral network to classify our data.
The resnet34 pretrained weights are downloaded. This is known as transfer learning; we do not want to start from scratch as it is very time consuming. Instead let's use a model that atleast recognizes a few pet species and then train it to recognize and classify all the species we want it to.

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

One cycle learning.

```python
learn.fit_one_cycle(4)
```

```python
learn.save('Initial Stage')
```

# Results
Let us now see how our model performed with different classes.

```python
interpretation = ClassificationInterpretation.from_learner(learn)
```

```python
interpretation.plot_top_losses(9, figsize=(15,11))
```

```python
interpretation.plot_confusion_matrix(figsize=(12,12))
```

```python
interpretation.most_confused(min_val=2)
```

# Making the model better 
## Unfreezing:
Inintially when we train the pretrained model by calling the fit or fit_one_cyle method on it, it does not run all of the layers. Instead only some of the layers are ran and thus the model runs quickly and doesn't overfit. By unfreezing, we are telling the model to run all the layers.

```python
learn.unfreeze()
```

```python
learn.fit_one_cycle(1)
```

Our error has increased instead of going further down. Why?

When we first ran the model, it kept the features learnt from the previous layer (from the pretrained weights) and then added some layers on top of those layers and then ran those layers, thus, learning on top the sophisticated features that the model had already learned.

But now that we're unfreezing the model. We try to relearn the sophisticated features for our pet classification that the imageNet learned for a different scenario. However, the features at the initial layers are pretty much the same in all cases (i.e mostly line and edge detection at the first few layers) so we do not need to relearn these; what actually needs relearning are the layers towards the middle and the end and make them localize to pet specie classification.

Our model takes all the layers and tries to relearn at the same pace, this causes the increase in the error rate.


```python
learn.load('Initial Stage')
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.unfreeze()
learn.fit_one_cycle(2, max_lr=slice(1e-50, 1e-45))
```

By changing just the learning rate on our initial model, we decreased the error rate from 0.066 to 0.062.
